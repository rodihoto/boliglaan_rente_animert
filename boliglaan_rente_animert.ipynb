```python
# Re-import necessary modules after code execution state reset
import pandas as pd
import matplotlib.pyplot as plt

# Re-create the dataset
data_bank = {
    "År": list(range(2015, 2026)),
    "DNB (%)": [3.25, 2.90, 2.80, 2.90, 2.95, 2.20, 1.99, 2.50, 4.83, 5.76, 5.44],
    "SpareBank 1 (%)": [2.90, 2.70, 2.65, 2.75, 2.80, 2.10, 1.90, 2.40, 4.70, 5.60, 5.30],
    "Sparebanken Sør (%)": [2.60, 2.40, 2.35, 2.45, 2.50, 2.00, 1.85, 2.35, 4.60, 5.50, 5.20]
}
df_banks = pd.DataFrame(data_bank)
df_banks
```

```python
# Plotting mortgage interest rates with custom colors using matplotlib
plt.figure(figsize=(12, 6))

# Plot each bank's line with distinct color
plt.plot(df_banks["År"], df_banks["DNB (%)"], label="DNB", color="darkblue", linewidth=2, marker="o")
plt.plot(df_banks["År"], df_banks["SpareBank 1 (%)"], label="SpareBank 1", color="green", linewidth=2, marker="o")
plt.plot(df_banks["År"], df_banks["Sparebanken Sør (%)"], label="Sparebanken Sør", color="red", linewidth=2, marker="o")

# Add labels and styling
plt.title("Boliglånsrenter i Norge (2015–2025)", fontsize=14)
plt.xlabel("År", fontsize=12)
plt.ylabel("Rente (%)", fontsize=12)
plt.grid(True)
plt.legend()
plt.xticks(df_banks["År"])
plt.tight_layout()
plt.show()
```
